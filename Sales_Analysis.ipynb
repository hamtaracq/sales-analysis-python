```python
import pandas as pd
p=pd.read_excel('sales.csv.xlsx')
COUNTRY_SALES = p.groupby("COUNTRY")["SALES"].sum()
print(COUNTRY_SALES)
```

```python
PRODUCT_AVG = p.groupby("PRODUCT")["SALES"].mean()
print(PRODUCT_AVG)
```

```python
COUNTRY_SALES=p.groupby("COUNTRY")["SALES"].sum()
print(COUNTRY_SALES.idxmax())
```

```python
import matplotlib.pyplot as plt
COUNTRY_SALES.plot(kind="bar")

plt.title("Sales by Country")
plt.ylabel("sales")
plt.show()
```
